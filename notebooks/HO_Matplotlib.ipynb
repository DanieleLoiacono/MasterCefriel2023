## Overview

![](images/pyplot.png)

*Matplotlib* is organized as a hierarchy. 
On top of the hierarchy is the matplotlib **state-machine environment** which is provided by the [matplotlib.pyplot](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.html#module-matplotlib.pyplot) module. 
At this level, simple functions are used to add plot elements (lines, images, text, etc.) to the **current axes** in the current figure.

To use the *Matplotlib* state-machine is necessary to import the *pyplot* module as:

```python
import matplotlib.pyplot as plt
```

The next level down in the hierarchy is the first level of the object-oriented interface, in which pyplot is used only for a few functions such as figure
creation, and the user explicitly creates and keeps track of the figure and axes objects. At this level, the user uses pyplot to create figures,
and through those figures, one or more axes objects can be created. These axes objects are then used for most plotting actions.

For even more control the pyplot level may be dropped completely, leaving a purely object-oriented approach.

At the lowest level, we have the **backends**, that are the components of *Matplotlib* in charge of rendering the output.
There are several backends, because there are several image formats and graphical libraries to render the output.
In particular **non-interactive backends** are used to write on a file, **interactive backends** are used to render on the screen.
A complete list of the available backends is described [here](https://matplotlib.org/tutorials/introductory/usage.html#backends).

## Anatomy of a figure

![](images/anatomy.png)

The components of the figure are represented in *Matplotlib* with the following objects:

### Figure
The **whole** figure that includes axes, titles, legend, labels, etc.
A figure can have several axes.

### Axes
The actual **plot**, that is are of the image where data is represented.
The Axes contains two (or three in the case of 3D) **Axis** objects.
This object is heavily used to work at the intermediate level mentioned before.

### Axis
They take care of setting the graph limits and generating the ticks (the marks
on the axis) and ticklabels (strings labeling the ticks).

### Artist
Basically everything you can see on the figure is an artist (even Figure, Axes, and Axis objects).
So, with this we refer to *text labels*, *line*, *legend*, and anything included in the figure.

## Basics
To generate a graph you can simply call one of the **pyplot** functions.
The simplest one is the **plot** function:

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.ylabel('Y')
plt.xlabel('X')
```

### Formatting 

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'gs-')
plt.axis([0, 6, 0, 20])
```

Different shape:

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'r-') 
```

Whit different line style:

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'r:') 
```

Plot several data together:

```python
import numpy as np #import numpy to generate numbers
import matplotlib.pyplot as plt

x = np.arange(0.0,5,0.2) #Create a sequence of number between 0 and 5 with 0.2 step
y1 = x
y2 = x**2 #x^2
y3 = x**3 #x^3

# red dashes, blue squares and green triangles
plt.plot(x, y1, 'r--') 
plt.plot(x, y2, 'bs')
plt.plot(x, y3, 'g^-')
plt.axis([0, 5.2, 0, 100])
```

Available option to change the plot style are [here](https://matplotlib.org/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D) and are listed with **setp()** function:

```python
lines = plt.plot([1,2,3]) #plot function return a reference to the lines plotted
plt.setp(lines)
```


### Dealing with categorical variables

## Multiple figures and axes

You can deal with multiple axes, using the pyplot state machines.

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.arange(0.0, 5.0, 0.1)
x2 = np.arange(0.0, 5.0, 0.02)
y1 = np.sin(2*x1)
y2 = np.cos(x2)

plt.figure(1) #create figure 1
plt.plot(x1,y1,'k')
plt.title('Figure 1')

plt.figure(2) #create figure 2 (now active)
plt.plot(x2,y2,'r-')
plt.title('Figure 2')

plt.figure(1) #go back to figure 1 (now active)
plt.title('New Figure 1')
```

Alternatively, you can work with the explicit references to figure and axes (intermediate level):

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.arange(0.0, 5.0, 0.1)
x2 = np.arange(0.0, 5.0, 0.02)
y1 = np.sin(2*x1)
y2 = np.cos(x2)

fig1, ax1 = plt.subplots() #create figure 1
fig2, ax2 = plt.subplots() #create figure 2
ax1.plot(x1,y1,'k') #create Figure 1
ax2.plot(x2,y2,'r-') #create Figure 2
ax1.set_title('Figure 1')
ax2.set_title('Figure 2')
```

You can also create a single figure with multiple axes:

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.arange(0.0, 5.0, 0.1)
x2 = np.arange(0.0, 5.0, 0.02)
y1 = np.sin(2*x1)
y2 = np.cos(x2)

plt.subplot(2,2,1) # or plt.subplot(121) - create a figure with a 1x2 grid of axes and make first axis active
plt.plot(x1,y1,'k')
plt.title('Axes 1')

plt.subplot(2,2,4) # or plt.subplot(122) - makes second axes active
plt.plot(x2,y2,'r-')
plt.title('Axes 2')
plt.suptitle('Figure with multiple axes')
```

Now multiple figures *and* multiple axes with explicit references:

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.arange(0.0, 5.0, 0.1)
x2 = np.arange(0.0, 5.0, 0.02)
y11 = np.sin(2*x1)
y12 = np.sin(0.5*x1)
y2 = np.cos(x2)

fig1, (ax11,ax12) = plt.subplots(1,2,sharey=True) #create 1x2 plot with shared axis
fig2, ax2 = plt.subplots() #create 1x1 plot

ax11.plot(x1,y11,'k')
ax12.plot(x1,y12,'k.-')
ax2.plot(x2,y2,'r--')
ax11.set_title('Axes1 - Fig1')
ax12.set_title('Axes2 - Fig1')
ax2.set_title('Figure2')
fig1.suptitle('Figure 1')
```

## Text and annotations

It is possible to add text in a figure:

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'r--')
plt.text(3, 2, 'Example')
```

We can also annotate the graphs:

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4, 5], [0.5, 1, 1.5, 8, 10], 'r')
plt.annotate('slope change', xy=(3, 1.5), xytext=(1, 7.5),arrowprops=dict(facecolor='green', shrink=5))
```
