# Seaborn

In this notebook, we will briefly introduce Seaborn, to show how it can be easily used to plot data. 

Seaborn can be imported with:

```python
import seaborn as sns
import matplotlib.pyplot as plt

#plt.style.use('seaborn-talk')
sns.set_context('talk')
plt.rcParams['figure.figsize']=(20,20)
```

Lets for example load iris data and plot something:

```python
import pandas as pd
import seaborn as sns

sns.set(rc={"figure.figsize": (16, 8)})
sns.set_context('talk')

data = pd.read_csv('data/iris.csv')
sns.lmplot(x='sepal_length',y='sepal_width', data=data)
plt.rcParams['figure.figsize']=(20,20)
```

Let remove the regression line:

```python
sns.lmplot(x='sepal_length',y='sepal_width', data=data, fit_reg=False)
```

We can change marks:

```python
sns.lmplot(x='sepal_length',y='sepal_width', data=data, fit_reg=False, markers='+')
```

We can plot different species with different colors:

```python
sns.lmplot(x='sepal_length',y='sepal_width', hue='species', data=data, fit_reg=False)
```

Or we can plot each species on a different subplot:

```python

sns.lmplot(x='sepal_length',y='sepal_width', col='species', data=data, fit_reg=False)
```

We can also generate more complex idioms very easily:


```python
sns.jointplot(x='sepal_length',y='sepal_width', data=data)
```

```python
sns.pairplot(data, hue="species")
```

It is also possible to customize the plots:

```python
sns.lmplot(x='sepal_length',y='sepal_width', hue='species', data=data, fit_reg=False, scatter_kws={'s':5})
```

```python
sns.lmplot(x='sepal_length',y='sepal_width', data=data, fit_reg=False, 
             scatter_kws={'s':60*data['petal_width']})
```
